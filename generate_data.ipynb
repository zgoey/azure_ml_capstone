```python
import pandas
df = pandas.read_csv("colornames.csv")
```

```python
# https://stackoverflow.com/a/26849064
green_df = df[df["name"].str.contains("Green|green", na=False)]
```

```python
pandas.set_option("display.max_rows", None)
display(green_df)
```

```python
def highlight(val):
    if val.startswith('#'):
        return 'background-color: %s' % val
    else:
        return ''
```

```python
s = green_df.style.applymap(highlight)
s
```
